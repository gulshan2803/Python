```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
column_name = ['Age' , 'Operation_Year' , 'Axil_Nodes' , 'Surv_Status'] 
df = pd.read_csv('haberman.csv' , header = None , names = column_name)
```

```python
df.head()
```

```python
df.tail()
```

```python
# Age
sns.FacetGrid(df, hue = 'Surv_Status' , size=4).map(sns.distplot, 'Age').add_legend();
plt.title('PLOT 1 CORRESPONDING TO AGE FEATURE')
plt.show();
```

1. Age 40-60 more likely to die, 2. Age less than 40 are more likely to survive.

```python
# Operation year
sns.FacetGrid(df, hue = 'Surv_Status' , size=4).map(sns.distplot, 'Operation_Year').add_legend();
plt.title('PLOT 1 CORRESPONDING TO OPERATION YEAR FEATURE')
plt.show();
```

It is clearly seen that the patients whose operation was done in between 60 to 65 died.

```python
# Axil Nodes
sns.FacetGrid(df, hue = 'Surv_Status' , size=4).map(sns.distplot, 'Axil_Nodes').add_legend();
plt.title('PLOT 1 CORRESPONDING TO AXILLARY NODES FEATURE')
plt.show();
```

It is clearly seen that patients having 0 axillary nodes will survive.
